```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 
from google.colab import drive
import matplotlib.image as mpimg
drive.mount('/content/drive')
%cd "/content/drive/My Drive/Colab Notebooks"
# import import_ipynb
import os
```

```
def get_all_images(folder, ext):

    all_files = []
    #Iterate through all files in folder
    for file in os.listdir(folder):
        #Get the file extension
        _,  file_ext = os.path.splitext(file)

        #If file is of given extension, get it's full path and append to list
        if ext in file_ext:
            full_file_path = os.path.join(folder, file)
            all_files.append(full_file_path)

    #Get list of all files
    return all_files

filepath1 = "/content/drive/My Drive/ready_data_6C/input/"
filepath2 = "/content/drive/My Drive/ready_data_6C/matrix_input/"

#List of all jps and tif files
run_file_images = get_all_images(filepath1, 'png')
Matrix_run_file_images = get_all_images(filepath1, 'png')

# tif_files = get_all_images(filepath2, 'tif')
```

```
# %pylab inline
# %matplotlib inline
# %pylab qt

for i in run_file_images:

  img=mpimg.imread(i)
  imgplot = plt.imshow(img)
  plt.show()
```
